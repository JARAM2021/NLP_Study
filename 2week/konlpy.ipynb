```
def line_preprocessing(line):
    if line[0] != " ":
        result_line = line
    else:
        for x in range(len(line)):
            if line[x] != " ":
                result_line = line[x:]
                break
    
    if line[0] == "\n":
        return None

    if result_line[-1] == "\n":
        result_line = result_line[:-1]

    s = result_line.split(" ")
    for x in s:
        if x != "":
            if x == s[0]:
                result_line = x
            else:
                result_line = result_line + " " + x

    if result_line == "":
        return None
    else:
        return result_line


f = open("/content/gdrive/MyDrive/난쏘공.txt", 'r')
pretreated_f = open("/content/gdrive/MyDrive/난쏘공_전처리.txt", 'w')



lines = f.readlines()
for line in lines:
    result_line = line_preprocessing(line)
    if result_line != None:
        pretreated_f.write(result_line+"\n")
        print(result_line)

f.close()
pretreated_f.close()
```

```
from konlpy.tag import Kkma
kkma = Kkma()

preprocessed_f = open("/content/gdrive/MyDrive/난쏘공_전처리.txt", "r")
mor_f = open("/content/gdrive/MyDrive/난쏘공_morphs.txt", "w")

lines = preprocessed_f.readlines()

for line in lines:
    mor = kkma.morphs(line)
    for word in mor:
        mor_f.write(word+"\n")


preprocessed_f.close()
mor_f.close()
```

```
preprocessed_f = open("/content/gdrive/MyDrive/난쏘공_전처리.txt", "r")
pos_f = open("/content/gdrive/MyDrive/난쏘공_pos.txt","w")

lines = preprocessed_f.readlines()

for line in lines:
    po = kkma.pos(line)
    for x in po:
        data = x[0]
        p = x[1]
        pos_f.write(data + " " + p + "\n")

preprocessed_f.close()
pos_f.close()
```

```
preprocessed_f = open("/content/gdrive/MyDrive/난쏘공_전처리.txt", "r")
nouns_f = open("/content/gdrive/MyDrive/난쏘공_nouns.txt", "w")

lines = preprocessed_f.readlines()

for line in lines:
    nouns = kkma.nouns(line)
    for n in nouns:
        nouns_f.write(n + "\n")


preprocessed_f.close()
nouns_f.close()
```

```
def Bubble_sort(li):
    for i in range(len(li)-1):
        for j in range(len(li)-i-1):
            if li[j][1]<li[j+1][1]:
                temp=li[j+1]
                li[j+1]=li[j]
                li[j]=temp
    return li

pos_f = open("/content/gdrive/MyDrive/난쏘공_pos.txt", "r")

lines = pos_f.readlines()

dic = {}

for line in lines:
    key = line.split(" ")[0]
    dic[key] = dic.get(key,0) + 1
    keys = sorted(dic.keys())

array = []

for key in keys:
    array.append([key, dic[key]])
    
# print(array)

result_array = Bubble_sort(array)

# print(result_array)

print(result_array[0:6])


pos_f.close()
```
