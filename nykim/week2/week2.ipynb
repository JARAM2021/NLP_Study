```
data = data.replace('. ', '.\n')
data = data.replace('\n\n','\n')
print(data)
sentences = data.split('\n')
```

## 2. morphs()

```
morphs = []
for sentence in sentences :
    morphs.append(kkma.morphs(sentence))

morphs
```

## 3. pos()

```
pos = []
for sentence in sentences :
    pos.append(kkma.pos(sentence))

pos
```

```
nouns = []
for sentence in sentences :
    nouns.append(kkma.nouns(sentence))

nouns
```
